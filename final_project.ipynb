```python
# Mendefinisikan kelas StudentManagementSystem
class StudentManagementSystem:
    # Metode inisialisasi kelas
    def __init__(self):
        # Inisialisasi list kosong untuk menyimpan data siswa
        self.students = []

    # Metode untuk menambahkan data siswa baru
    def add_student(self):
        # Menampilkan header untuk penambahan data siswa
        print("\n=== Tambah Data Siswa ===")
        # Meminta input nama siswa
        nama = input("Masukkan nama: ")
        # Loop untuk memastikan input NIM valid
        while True:
            try:
                # Meminta input NIM
                nim = input("Masukkan NIM: ")
                # Memeriksa apakah NIM sudah ada dalam daftar siswa
                if any(student['nim'] == nim for student in self.students):
                    print("NIM sudah terdaftar!")
                    continue  # Kembali ke awal loop jika NIM duplikat
                break  # Keluar dari loop jika NIM valid
            except ValueError:
                # Menangkap error jika input bukan angka
                print("NIM harus berupa angka!")

        # Meminta input jurusan
        jurusan = input("Masukkan jurusan: ")
        # Loop untuk memastikan input nilai valid
        while True:
            try:
                # Meminta input nilai dan mengkonversi ke float
                nilai = float(input("Masukkan nilai: "))
                # Memeriksa apakah nilai berada dalam rentang 0-100
                if 0 <= nilai <= 100:
                    break  # Keluar dari loop jika nilai valid
                print("Nilai harus antara 0-100!")
            except ValueError:
                # Menangkap error jika input bukan angka
                print("Nilai harus berupa angka!")

        # Membuat dictionary untuk menyimpan data siswa
        student = {
            'nama': nama,
            'nim': nim,
            'jurusan': jurusan,
            'nilai': nilai
        }
        # Menambahkan data siswa ke dalam list
        self.students.append(student)
        # Menampilkan pesan sukses
        print("Data siswa berhasil ditambahkan!")

    # Metode untuk menampilkan semua data siswa
    def display_students(self):
        # Memeriksa apakah ada data siswa
        if not self.students:
            print("\nBelum ada data siswa!")
            return

        # Menampilkan header untuk data siswa
        print("\n=== Data Semua Siswa ===")
        # Loop untuk menampilkan data setiap siswa
        for student in self.students:
            print(f"\nNama: {student['nama']}")
            print(f"NIM: {student['nim']}")
            print(f"Jurusan: {student['jurusan']}")
            print(f"Nilai: {student['nilai']}")
        print("\n")

    # Metode untuk mencari siswa berdasarkan NIM
    def search_student(self):
        # Memeriksa apakah ada data siswa
        if not self.students:
            print("\nBelum ada data siswa!")
            return

        # Meminta input NIM yang dicari
        nim = input("\nMasukkan NIM yang dicari: ")
        found = False
        # Loop untuk mencari siswa dengan NIM yang sesuai
        for student in self.students:
            if student['nim'] == nim:
                # Menampilkan data siswa jika ditemukan
                print(f"\nNama: {student['nama']}")
                print(f"NIM: {student['nim']}")
                print(f"Jurusan: {student['jurusan']}")
                print(f"Nilai: {student['nilai']}")
                found = True
                break

        # Menampilkan pesan jika siswa tidak ditemukan
        if not found:
            print("Siswa dengan NIM tersebut tidak ditemukan!")

    # Metode untuk menghitung rata-rata nilai semua siswa
    def calculate_average(self):
        # Memeriksa apakah ada data siswa
        if not self.students:
            print("\nBelum ada data siswa!")
            return

        # Menghitung total nilai semua siswa
        total_nilai = sum(student['nilai'] for student in self.students)
        # Menghitung rata-rata nilai
        rata_rata = total_nilai / len(self.students)
        # Menampilkan hasil rata-rata dengan 2 angka desimal
        print(f"\nRata-rata nilai semua siswa: {rata_rata:.2f}")

    # Metode untuk menampilkan dan mengelola menu utama
    def main_menu(self):
        # Loop utama program
        while True:
            # Menampilkan opsi menu
            print("\n=== Sistem Manajemen Siswa ===")
            print("1. Tambah data siswa")
            print("2. Tampilkan semua data siswa")
            print("3. Cari siswa berdasarkan NIM")
            print("4. Hitung rata-rata nilai")
            print("5. Keluar")

            # Meminta input pilihan menu dari pengguna
            choice = input("Pilih menu (1-5): ")

            # Menjalankan metode yang sesuai berdasarkan pilihan pengguna
            if choice == '1':
                self.add_student()
            elif choice == '2':
                self.display_students()
            elif choice == '3':
                self.search_student()
            elif choice == '4':
                self.calculate_average()
            elif choice == '5':
                # Menampilkan pesan penutup dan keluar dari program
                print("\nTerima kasih telah menggunakan program ini!")
                break
            else:
                # Menampilkan pesan error jika pilihan tidak valid
                print("\nPilihan tidak valid!")

# Blok utama program
if __name__ == "__main__":
    # Membuat instance dari kelas StudentManagementSystem
    sms = StudentManagementSystem()
    # Menjalankan menu utama
    sms.main_menu()
```
